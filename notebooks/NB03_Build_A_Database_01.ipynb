````python
import os
import sqlite3


import numpy as np
import pandas as pd
````

````python
DATA_FOLDER = os.path.join('../data/')

RAW_FOLDER = os.path.join(DATA_FOLDER, 'raw')
````

````python
conn = sqlite3.connect(os.path.join(DATA_FOLDER, 'football.db'))
````

### Table 1: Teams

````python
df_teams = pd.read_csv('../data/raw/text_full_final.csv')
df_teams_xg = pd.read_csv('../data/raw/team_xg.csv')

````

````python
df_teams_xg['League'] = df_teams_xg['Leagues']
df_teams_xg = df_teams_xg.drop('Leagues', axis='columns')
````

````python
df_teams = pd.merge(df_teams, df_teams_xg, on=['Teams', 'League'], how='inner')
df_teams
````

````python
df_teams.info()
````

````python
def extract_team_id(url):
    parts = url.split('/')
    return parts[4]
````

````python
df_teams['Team ID'] = df_teams['Links'].apply(extract_team_id)
df_teams
````

````python
df_teams.to_sql('teams', conn, if_exists='replace', index=False)
````

````python
pd.read_sql('SELECT * FROM teams LIMIT 10', conn)
````

### Table 2: Players

````python
df_players = pd.read_csv('../data/raw/players_data_full.csv')
df_player_xg = pd.read_csv('../data/raw/players_xg_data.csv')
````

````python
df_players
````

````python
df_player_xg
df_player_xg.rename(columns = {'Aerial Duels Won':'Expected goals per shot'}, inplace = True)
````

````python
df_players = pd.merge(df_players, df_player_xg, on=['Player Name'], how='inner') #LUEGO CAMBIAR
df_players.replace('-', 0, inplace=True)
df_players.rename(columns = {'Aerial Duels Won_y':'Expected goals per shot'}, inplace = True)
df_players
````

````python
df_players['Team ID'] = df_players['Team Link'].apply(extract_team_id)
df_players
````

````python
df_players.to_sql('players', conn, if_exists='replace', index=False)
````

Adding the corresponding team name to each player by the team id from the link

````python
df_temporary = pd.read_sql(
'''
SELECT 
    players.*, 
    teams.[Teams]
FROM 
    players
INNER JOIN 
    teams
ON 
    players.[Team ID] = teams.[Team ID]
''', conn)
````

````python
df_temporary.to_sql('players', conn, if_exists='replace', index=False)
````

````python
pd.read_sql('SELECT * FROM players LIMIT 5', conn)
````
